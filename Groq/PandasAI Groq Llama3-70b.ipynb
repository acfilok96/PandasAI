```python
from google.colab import userdata
GROQ_API_KEY = userdata.get('GROQ_API_KEY')
```

```python
import pandas as pd
from pandasai import SmartDataframe
from langchain_groq.chat_models import ChatGroq
from langchain_community.llms import Ollama
import sqlite3
import os
```

```python
llm = ChatGroq(model_name = "llama3-70b-8192",
               api_key = GROQ_API_KEY)
```

```python
df = pd.read_excel('/content/33c4f9b8-data.xlsx')
```

```python
df = SmartDataframe(df,
                    config = {"llm": llm})
```

```python
response1 = df.chat('Which are the 5 happiest countries?')
print(response1)
```

```python
response2 = df.chat('What is the sum of the GDPs of the 2 happiest countries?')
print(response2)
```
